```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

random.seed(42)
np.random.seed(42)

# Config
n_records = 10000
start_date = datetime(2023, 1, 1)

categories = {
    "Groceries": ("Whole Foods", "Trader Joe's", "Costco", "Aldi"),
    "Dining": ("Starbucks", "McDonald's", "Chipotle", "DoorDash", "Uber Eats"),
    "Transport": ("Uber", "Lyft", "Shell Gas", "BP", "Metro Card"),
    "Entertainment": ("Netflix", "Spotify", "Steam", "AMC Theaters"),
    "Shopping": ("Amazon", "Zara", "H&M", "Nike", "ASOS"),
    "Utilities": ("Electric Bill", "Water Bill", "Internet Bill", "Phone Bill"),
    "Health": ("CVS Pharmacy", "Gym Membership", "Doctor Visit", "Dentist"),
}

rows = []
for i in range(n_records):
    category = random.choice(list(categories.keys()))
    merchant = random.choice(categories[category])
    amount = round(random.uniform(3.5, 400.0), 2)
    date = start_date + timedelta(days=random.randint(0, 730))
    rows.append({
        "transaction_id": f"TXN{str(i+1).zfill(5)}",
        "date": date.strftime("%Y-%m-%d"),
        "merchant": merchant,
        "category": category,
        "amount": amount,
        "currency": "USD",
        "status": random.choice(["completed", "completed", "completed", "pending", "failed"]),
    })

df = pd.DataFrame(rows)


df.loc[df.sample(frac=0.02).index, "amount"] = None   # missing amounts
df.loc[df.sample(frac=0.01).index, "category"] = None # missing categories
df = pd.concat([df, df.sample(100)])                    # duplicate rows

print(f"Generated {len(df)} records")
df.head()
```

```python
# Save to Lakehouse Bronze folder
df.to_csv("/lakehouse/default/Files/bronze/transactions/transactions_raw.csv", index=False)
print("File saved to Bronze layer")
```
